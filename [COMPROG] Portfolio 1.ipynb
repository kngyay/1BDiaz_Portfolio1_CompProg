```python
# PORTFOLIO 1: LIST INPUT
# CONCEPT: LAB DIGITAL CALCULATOR 

# During a laboratory activity, the preparation of a reagent solution is one of the most crucial steps in beginning the experiment. 
# In a setting where accuracy and precision are indespensible, an automatic calculator for the number of reagent powder needed becomes a 
# helpful tool for efficiency and correctness. 
```

```python
ReagentName=input("Reagent:")
molarity=float(input("M of Solution Needed:"))
volume=float(input("Volume (L) of Solution Needed:"))
mmass=float(input("Molar Mass of Reagent:"))
GramsReagent=float(molarity*volume*mmass)
print("You need",GramsReagent, "g of",ReagentName,"for",volume,"L of",molarity,"M solution.")
```
